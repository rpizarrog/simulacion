## Construir el gráfico de qqplot

El gráfico de comparación de cuantiles, también llamado, *qqplot*, representa los cuantiles de los datos comaprado contra los cuantiles teóricos que tendría si siguiese una distribución normal con una media y desviación estándar igual.

Si los datos siguen una distribución normal, se alinearán cerca de la diagonal del gráfico. Cuanto más se alejen, menos probable será que nuestros datos sigan una distribución normal.

```python
# Crear el QQ-Plot
sm.qqplot(np.array(datos), line='q')

# Configurar el gráfico
plt.title('QQ-Plot')
plt.xlabel('Cuantiles teóricos de la distribución normal')
plt.ylabel('Cuantiles de los datos')

# Mostrar el gráfico
plt.show()
```

No se observa una alineación consistente con la línea roja que representa la distribución normal teórica simulada a partir de los mismos valores estadístcos de media y desviación estándar.

Los datos no provienen de una distribución normal

Esta prueba enriquece, fortalece y corrobora las pruebas de *Kolmogorov Smirnov* y *Shapiro Wilk*.
